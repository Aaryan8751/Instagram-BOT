```python
#import of some libraries
from bs4 import BeautifulSoup
from selenium import webdriver
import sys
from selenium.common.exceptions import NoSuchElementException
import time
```

```python
#some important functions needed time to time

#Searchbartext helps us to write the text in the search bar.It first find the search bar
#and then click on the search bar and the send the text by finding the searchbartext element by classname
def Searchbartext(text):
    search_bar = driver.find_element_by_class_name("eyXLr.wUAXj")
    search_bar.click()
    search_bar_text = driver.find_element_by_class_name("XTCLo.x3qfX")
    search_bar_text.send_keys(text)

#clearSearch helps us to erase the text from the seachbar and close the searchbar
def clearSearch():
    search_bar_text = driver.find_element_by_class_name("XTCLo.x3qfX")
    search_bar_text.clear()
    close = driver.find_element_by_class_name('aIYm8.coreSpriteSearchClear')
    close.click()

#closefollowerswindow helps us to close the followers window by find the class of close and by clicking on it
def closefollowerswindow():
    close_followers = driver.find_elements_by_class_name("wpO6b")[1]
    close_followers.click()
    
```

# 5. Like/Unlike posts 

```python
# calling function Searchbartext and filling dilsefoodie in the search bar
Searchbartext('dilsefoodie')
```

```python
# stopping for 3 sec
time.sleep(3)
```

```python
# getting the elements of the posts and keeping the 30 in the list
posts_of_dilsefoodie = driver.find_elements_by_class_name("v1Nh3.kIKUG._bz0w")[:30]
```

```python
# stopping for 2 sec
time.sleep(2)
```

### 5.1 Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
# applying for loop on the elements
for posts in posts_of_dilsefoodie:
    # clicking on the post on by one
    posts.click()
    # implicitly waiting for 10 sec
    driver.implicitly_wait(10)
    # getting the element of like button 
    like = driver.find_element_by_class_name("fr66n")
    # getting the data of html which is inside the tag
    data = BeautifulSoup(like.get_attribute('innerHTML'),'html.parser')
    # saving the liked or not data
    liked_or_not = data.svg['aria-label']
    # if the post is already liked then liked_or_not will have unlike in it so we will print the message and close the post
    # window
    if liked_or_not=='Unlike':
        print("You have already liked the top 30 Posts")
        close = driver.find_element_by_class_name("Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG")
        close.click()
        break
    # liking the post if not already liked
    like.click()
    # closing the post window by obtaining the close element
    close = driver.find_element_by_class_name("Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG")
    close.click()
    # implicitly waiting for 10 sec
    driver.implicitly_wait(10)
    
```

### 5.2 Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
# applying for loop on the elements
for posts in posts_of_dilsefoodie:
    # clicking on the post on by one
    posts.click()
    # getting the element of like button 
    like = driver.find_element_by_class_name("fr66n")
    # getting the data of html which is inside the tag
    data = BeautifulSoup(like.get_attribute('innerHTML'),'html.parser')
    # saving the liked or not data
    liked_or_not = data.svg['aria-label']
    # if the post is already unliked then liked_or_not will have like in it so we will print the message and close the post
    # window
    if liked_or_not=='Like':
        print("You have already Unliked the top 30 Posts")
        close = driver.find_element_by_class_name("Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG")
        close.click()
        break
    # unliking the post if not already liked
    like.click()
    # closing the post window by obtaining the close element
    close = driver.find_element_by_class_name("Igw0E.IwRSH.eGOV_._4EzTm.BI4qX.qJPeX.fm1AK.TxciK.yiMZG")
    close.click()
    # implicitly waiting for 10 sec
    driver.implicitly_wait(10)
    
```

```python
def followers_equal_to_500():
    # swiping down the followers window so that we obtaing the class wo9IH in the html 
    swipedown = driver.find_element_by_class_name("oMwYe")
    swipedown.click()
    # creating a helper list
    helper_list = []
    while len(helper_list) < 500:
        # getting all the follower elements and storing it in the list 
        helper_list = driver.find_elements_by_class_name("wo9IH")
        # swiping down the followers window
        swipedown = driver.find_element_by_class_name("oMwYe")
        swipedown.click()
        # waiting 
        driver.implicitly_wait(10)
        time.sleep(2)
    # returning the helper list
    return helper_list
```

```python
# saving the list of followers of foodtalkindia by calling the function followers_equal_to_500()
li_usernames_of_followers_of_foodtalkindia = followers_equal_to_500()
```

```python
# printing the 500 followers of foodtalkindia
for usernames in li_usernames_of_followers_of_foodtalkindia[:500]:
    print(usernames.text.split('\n')[0])
```

### 6.2 Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

```python
# calling function Searchbartext and filling foodtalkindia in the search bar
Searchbartext('foodtalkindia')
```

```python
# getting the element of foodtalkindia username and opening the foodtalkindia instagram handle
foodtalkindia = driver.find_element_by_class_name("yCE8d")
foodtalkindia.click()
```

```python
# stoping for 2 sec and then getting the mutual followers element and clicking on it
time.sleep(2)
mutualfollowers = driver.find_element_by_class_name('tc8A9')
mutualfollowers.click()
```

```python
# stoping for 2 sec and then getting the see all followers element and clicking on it
time.sleep(2)
see_all_followers = driver.find_element_by_class_name('sqdOP.yWX7d.y3zKF.ZIAjV')
see_all_followers.click()
```

```python
# swiping down the mutual followers window so that we may obtain wo9IH class in the html
swipedown = driver.find_element_by_class_name("oMwYe")
swipedown.click()
# creating a list of the usernames that i am following and who follow the foodtalkindia
li_of_following = []
trap = ''
while trap!='Follow':
    # getting the elements of insta handles and storing in li_usernames_of_followers_of_foodtalkindia
    li_usernames_of_followers_of_foodtalkindia = driver.find_elements_by_class_name("wo9IH")
    # running a for loop on the insta handles stored in li_usernames_of_followers_of_foodtalkindia
    for ele in li_usernames_of_followers_of_foodtalkindia:
        # getting the text of one insta handle and spliting on \n
        li_helper = ele.text.split('\n')
        # assigning trap li_helper[-1] which is Follow or Following
        trap = li_helper[-1]
        # if trap is equal to following the we append the username of the insta handle to the li_of_following else we break
        if trap == 'Following':
            li_of_following.append(li_helper[0])
        else:
            break
        
```

```python
# printing all the usernames that i follow and those who follow foodtalkindia
for username in li_of_following:
    print(username)
```

# Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly -
### If You have already seen the story.
### Or The user has no story.
### Or View the story if not yet seen.

```python
# calling function Searchbartext and filling coding.ninjas in the search bar
Searchbartext('coding.ninjas')
```

```python
# getting the element of the story of coding ninjas
story_details = driver.find_element_by_class_name('z556c')
# getting the inner data of the html tag
data = BeautifulSoup(story_details.get_attribute('innerHTML'),'html.parser')
# getting the check_story integer
check_story = int(data.div['tabindex'])
# getting the viewed or not integer
viewed_or_not = int(data.div.canvas['height'])
# if the check_story is equal to 0 then the user has story else there is no story no if the check story is 0 and if
# viewed_or_not is equal to 53 then we have not seen the story else we have already seen the story
if check_story == 0:
    if viewed_or_not == 53:
        print('View the story if not yet seen.')
    else:
        print('You have already seen the story')
else:
    print('The user has no story.')
```
